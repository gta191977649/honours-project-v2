# Graph For arousal

```python
song_id = 123

import pandas as pd

a = pd.read_csv('./dataset/arousal.csv')

a = a.loc[a['song_id'] == song_id,'15000' : '44500']

time_interval = a.columns.values
acrosual = a.values[0]


#Valance
v = pd.read_csv('./dataset/valence.csv')

v = v.loc[v['song_id'] == song_id,'15000' : '44500']

valance = v.values[0]

df = pd.DataFrame({
    'A': acrosual,
    'V': valance
},index=time_interval)
df.plot.line()
```

## Dataset Distributions

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./dataset/labeled.csv')

v = data['valance']
a = data['arousal']
plt.scatter(v,a,s=5)
plt.title('V & A Distributions')
plt.show()
```

## Feature graph

```python
song_id = 10

import pandas as pd

file = "./dataset/memd_set/{}.csv".format(song_id)
data = pd.read_csv(file,sep=";")

featureSet = {}



for col in data.columns.values:
    featureSet[col] = data[col].tolist()

    
df = pd.DataFrame(featureSet,index=featureSet["frameTime"])
df.plot.line()
```
