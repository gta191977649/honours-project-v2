# Feature Ranking Benchmark (Random Forest)

```python
import numpy as np
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("../rfe_benchmark_random_forest_av.csv")

# Sample data
x = np.array(data['n_of_feature'])
ay = np.array(data['a'])


plt.plot(x,ay,'.')
vy = np.array(data['v'])
plt.plot(x,vy,'.')

z = np.polyfit(x, ay,6)
p = np.poly1d(z)
plt.plot(x,p(x),"-")


#slope = p(1)
print("y={}".format(p))




z = np.polyfit(x, vy,6)
p = np.poly1d(z)
plt.plot(x,p(x),"-")

#slope = p(1)
print("y={}".format(p))

plt.legend(['Arousal','Valance'])
```
